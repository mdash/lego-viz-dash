```python
themes_data['parent_id_final'] = themes_data.apply(lambda x: x['id'] if pd.isnull(x['parent_id']) else x['parent_id'],axis=1)
```

```python
themes_data.parent_id_final.nunique()
```

```python
themes_data['parent_name'] = themes_data[['parent_id_final','id']].merge(themes_data[['id','name']],left_on='parent_id_final',right_on='id',how='inner')['name']
```

```python
sets_data['parent_theme_name'] = sets_data.merge(themes_data,left_on='theme_id',right_on='id',how='left')['parent_name']
```
